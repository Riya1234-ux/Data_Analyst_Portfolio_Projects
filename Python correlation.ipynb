```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']= (12,8)

df= pd.read_csv(r"C:\Users\user\Downloads\movies.csv")
```

```python
df.head()
```

```python
# check for missing values
for col in df.columns:
    pct_missing= np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
df['yearcorrect']= pd.to_numeric(df['released'].str.extract(r'(\d{4})')[0],errors='coerce')
df['yearcorrect']= df['yearcorrect'].fillna(0).astype(int)
df.head()
```

```python
df= df.sort_values(by=['gross'],inplace=False,ascending=False)
df
```

```python
pd.set_option('display.max_rows',None)
df
```

```python
df.drop_duplicates()
```

```python
#Scatter plot with budget v/s gross
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Plot')
plt.xlabel('Gross')
plt.ylabel('Budget')
plt.show()
```

```python
sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
#looking at correlation
df.corr(method='pearson',numeric_only=True)
```

```python
#high corr btw budget and gross
corr_matrix= df.corr(method='pearson',numeric_only=True)
sns.heatmap(corr_matrix,annot=True)
plt.title('Correlation Plot')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
plt.show()
```

```python
#look at company
df_numerized= df.copy()
df_numerized=df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
correlation_matrix= df_numerized.corr(method='pearson',numeric_only=True)
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Plot')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
plt.show()
```

```python
corr_mat= df_numerized.corr(numeric_only=True)
corr_pairs= corr_mat.unstack()
corr_pairs
sorted_pairs= corr_pairs.sort_values()
high_corr= sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
